```python
# Download needed files
import nltk
nltk.download('all')
```

```python
# Import for NLP Model
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize , sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
data
```

```python
filters = set(stopwords.words('english'))
ps = PorterStemmer()
```

```python
reviews = []
for i in range(len(data)):
  clean = re.sub('[^a-zA-Z]',' ',data['reviewText'][i])
  clean = clean.lower()
  clean = word_tokenize(clean)
  clean = [ps.stem(word) for word in clean if word not in filters ]
  clean = ' '.join(clean)
  reviews.append(clean)
reviews
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(reviews)
print(x)
```

```python
y = data['Positive']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(learning_rate = 0.2 , n_estimators = 200)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
